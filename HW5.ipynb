다형성:  상속 관계내의 다른 클래스들의 인스턴스들이 같은 멤버함수 호출에 대해 각각 다르게 호출하도록 하는 기능

##  [3번 문제]<br >
다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)<br >

- 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.<br >

count: Counter 인스턴스가 지니는 초기 정수 값<br >
step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)<br >
'>>>' c = Counter(10) <br >
'>>>' d = Counter(10, 2) <br >
- 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.<br >

[Count (step: 1)] 10 <br >
'>>>' print(d) <br >

- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.<br >

 - 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.<br >

'>>>' c() <br >
'>>>' d() <br >
'>>>' print(c) <br >
[Count (step: 1)] 12 <br >
'>>>' print(d) <br >
[Count (step: 2)] 14 <br >
- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.<br >

'>>>' c = c + 5 <br >
'>>>' d = d - 5 <br >
'>>>' print(c) <br >
[Count (step: 1)] 17 <br >
'>>>' print(d) <br >
[Count (step: 2)] 9 <br >
- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.<br >

'>>>' print(c) > 10 <br >
True <br >
'>>>' print(d) > 10 <br >
False <br >
'>>>' print(c) < 10 <br >
False <br >
'>>>' print(d) < 10 <br >
True <br >
'>>>' print(c) == 17 <br >
True <br >
'>>>' print(d) != 9 <br >
False <br >

```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
    def __str__(self):
        result = "[Counter(step: %d)] %d" %(self.step, self.count)
        return result

    def incr(self):
        self.count = self.count+ self.step
    def __call__(self):
        self.incr()
    def __add__(self, other):
        self.count=self.count + other
        return self.count
    def __sub__(self, other):
        self.count=self.count - other
        return self.count
        
    def __lt__(self, y):
        return self.count < y
    def __eq__(self, y):
        return self.count == y
    def __ne__(self, y):
        return self.count != y
    
print("조건1,2")           
a=Counter(10)
b=Counter(10,5)
print(a)
print(b)
print("조건3")
a.incr()
b.incr()
print(a)
print(b)
print("조건4")
a()
b()
print(a)
print(b)
print("조건5")
a=a+2
b=b-3
print(a)
print(b)
print("조건6")
print(a>10)
print(a<10)
print(b>10)
print(b<10)
print(a==17)
print(b!=16)
```

요구사항 1번은 생성자에서 count랑 step을 받게 하였고
step은 주지 않으면 1로 하게 하였다.
str메소드는 결과를 %d를 사용하여 step과 count를 출력하게하였다
incr메소드는 count에 step을 계속해서 더해주었다
__call__을 이용하여 객체를 직접호출하게 하였다. 호출하면 내부적으로 incr()메소드를 호출하게 하였다.
산술연산은 각각 __add__, __sub__를 사용하였다
관계 연산은 __lt__, __eq__, __ne__를 사용하였다.

## [4번 문제]<br >
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.<br >

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

init함수는 생성자로 리스트를 받아서 자기자신 리스트를 생성 후 eliminate_duplicate함수로 넘겨주고 있고 변수 값 초기화가 가능한 함수이다.
str함수는 print 예약어나 str()내장함수 호출에 대응되는 메소드이고 Myset을 출력하게 하는 함수이다
eliminate_duplicate 함수는 인스턴스 객체와 무관하게 클래스 이름 공간에 존재하는 메소드로서 클래스 이름을 이용하여 직접 호출할 수 있는 메소드이다.  리스트를 받아서 중복된 것이 있으면 제거하고 겹치치않는 값으로만 만든다.
